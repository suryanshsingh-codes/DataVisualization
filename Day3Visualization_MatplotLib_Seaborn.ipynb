# **MatPlotLib**

**Importing**

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
x = np.linspace(0,5,11) # generating an array using numPy
x
```

```python
y = x **3 #creating another array by broadcasting previous array
y
```

```python
plt.plot(x,y) #plot x and y
plt.title("My First Plot")
plt.xlabel("X-NP Array as X label")
plt.ylabel("Y-NP Array as Y label")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1) # divide the whole canva into 2 * 2 and selecting part 1
plt.plot(x**2,y**2) 

plt.subplot(2,2,2) # divide the whole canva into 2 * 2 and selecting part 2
plt.plot(x-3,y+3)

plt.subplot(2,2,3)# divide the whole canva into 2 * 2 and selecting part 3
plt.plot(x,y)

plt.subplot(2,2,4)# divide the whole canva into 2 * 2 and selecting part 4
plt.plot(y,x)
```

**Object Oriented Way**

```python
fig=plt.figure() # this will generate a canvas

axis1=fig.add_axes([0.1,0.2,0.6,0.5]) #create a axis(Graph 1) and add axes and these values are from 0 to 1: which are The dimensions (left, bottom, width, height) of the axes.
axis1.plot(x,y)

axis2=fig.add_axes([0.9,0.2,0.6,0.75])
axis2.plot(x,y)
```

```python
fig=plt.figure(figsize=(7,8),dpi=(70)) # creating a canvas whose FIGSIZE is its size and DPI is the pixel density

axis1=fig.add_axes([0.1,0.2,0.6,0.75])
axis1.plot(x,y)
```

```python
fig=plt.figure(figsize=(7,8),dpi=(70))
axis1=fig.add_axes([0.1,0.2,0.6,0.75])
axis1.plot(x,y)
plt.savefig("FirstSavedPlot.png")
```

*Import the values: Seaborn*

**Loading The Dataset**

```python
df = sns.load_dataset('tips')
df
```

```python
df['size'].unique() # check the unique values
```

**Distplot/ Histplot**

```python
sns.histplot(df['total_bill'])
sns.histplot(df['total_bill'], bins=10) # bins are the boxes drawn for the value density, as mentioned in notes
sns.histplot(df['total_bill'], bins=10, kde=True)
```

```python
#plotting two graphs on a canvas

plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 20,kde = True)

plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

# **Regression Plot**

**lmplot()**

```python
tips = sns.load_dataset("tips")
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':50})
```

```python
tips.groupby('day')['tip'].mean().iplot(kind = 'bar')
```

```python
tips.iplot(kind='scatter',x ='total_bill',y = 'tip',mode = 'markers')
```

```python
tips[['total_bill','day']].iplot(kind = 'box',x = 'total_bill',y = 'day')
```
